## Exp no 9 - Epoch 50


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
import cv2
from glob import glob
import os
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.metrics import classification_report
from sklearn import metrics
import pandas as pd

seed = 182510
np.random.seed(seed)
tf.random.set_seed(seed)

def procImages(dataDir, labelDictionary, width, height):
    
    x = [] # will store images as arrays
    y = [] # store labels
    # list folders in directory
    directory = os.listdir(dataDir)
     
    # for each folder (train and validation) 
    for label in directory:
        
        # add class label to label dictionary
        if label not in labelDictionary:
            labelDictionary[label] = len(labelDictionary)
        
        # create full path for image directory (append absolute path and 
        # image directory path)
        sourceImages = os.path.join(dataDir, label)
        images = os.listdir(sourceImages)
        # for each image in directory, 
        for image in images:
            
            # read the image from file, resize and add to a list
            full_size_image = cv2.imread(os.path.join(sourceImages, image))
            x.append(cv2.resize(full_size_image, (width,height), 
                                                         interpolation=cv2.INTER_CUBIC))
            # add the class label to y
            y.append(label)

    return np.array(x), np.array(y)


def plotAccLoss(H, NUM_EPOCHS):

    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.show()


def main():
    # dimensions of our images.
    width, height = 64, 64
    K_fold = 5
    training_acc  = []
    validation_acc = []
    training_loss = []
    validation_loss = []
    K_precision = []
    K_recall = []
    K_fscore = []
    K_validation_acc = []

    for i in range(K_fold):
        print('\nExecuting Fold: '+ str(i+1))
        print('Current Seed: ', np.random.get_state()[1][0])
        picFolderName = "Fold"+ str(i+1)
        train_data_dir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_training_set',picFolderName)
        validation_data_dir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_test_set',picFolderName)
        numTrainingSamples = 572
        numValidationSamples = 140
        NUM_EPOCHS = 50
        batchSize = 35

        classifier = Sequential()

        classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Conv2D(64, (3, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Flatten())
        classifier.add(Dense(units = 256, activation = 'relu', kernel_regularizer=regularizers.l2(l=0.01)))
        # classifier.add(Dropout(0.5))
        classifier.add(Dense(units = 10, activation = 'softmax')) # number of classes

        classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

        # This dictionary will contain all labels with an associated int value
        labelDictionary ={}
        # The procImages function will read all image data from a folder
        # convert it to a Numpy array and add to a list
        # It will also add the label for the image, the folder name
        trainImages, trainLabels = procImages(train_data_dir, labelDictionary, width, height)
        valImages, valLabels = procImages(validation_data_dir, labelDictionary, width, height)

        # Map string label values to integer values. 
        trainLabels = (pd.Series(trainLabels).map(labelDictionary)).values
        valLabels = (pd.Series(valLabels).map(labelDictionary)).values

        trainLabels_values = np.max(trainLabels) + 1 
        trainLabels_onehot = np.eye(trainLabels_values)[trainLabels]
        valLabels_values = np.max(valLabels) + 1
        valLabels_onehot = np.eye(valLabels_values)[valLabels]

        train_datagen = ImageDataGenerator(
              rescale=1. / 255,
              shear_range=0.2,
              zoom_range=0.2,
              rotation_range = 20,
              horizontal_flip=True)

        train_generator = train_datagen.flow(
              trainImages, trainLabels_onehot, batch_size=batchSize, seed = seed)

        test_datagen = ImageDataGenerator(rescale=1. / 255)


        validation_generator = test_datagen.flow(valImages, valLabels_onehot, batch_size=batchSize, shuffle = False, seed = seed)

        history = classifier.fit(
            train_generator,
            steps_per_epoch=numTrainingSamples // batchSize,
            epochs=NUM_EPOCHS,
            validation_data=validation_generator,
            validation_steps=numValidationSamples // batchSize)
        
        training_acc.append(history.history["accuracy"])
        validation_acc.append(history.history["val_accuracy"])
        training_loss.append(history.history["loss"])
        validation_loss.append(history.history["val_loss"])

        plotAccLoss(history, NUM_EPOCHS)

        classifier_json= classifier.to_json()
        with open("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp10.Fold_"+str(i+1)+"json", "w") as json_file:
              json_file.write(classifier_json)
          
        # serialize the model weights to a HDF5 file
        classifier.save_weights("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp10.Fold_"+str(i+1)+".h5")

        val_trues = valLabels
        validation_generator.reset()
        y_pred = classifier.predict(validation_generator, steps = numValidationSamples // batchSize)
        val_preds = np.argmax(y_pred, axis=1)

        labels = labelDictionary.keys()
        report = classification_report(val_trues, val_preds, output_dict=True)
        print(classification_report(val_trues, val_preds, target_names = labels))
        K_precision.append(report['weighted avg']['precision'])
        K_recall.append(report['weighted avg']['recall'])
        K_fscore.append(report['weighted avg']['f1-score'])
        K_validation_acc.append(report['accuracy'])
        print(metrics.confusion_matrix(val_trues, val_preds))

    print('\n')
    print('Average precision: ', np.mean(K_precision))
    print('Average recall: ', np.mean(K_recall))
    print('Average fscore: ', np.mean(K_fscore))
    print('Report accuracy: ', np.mean(K_validation_acc))
    print("Average Training accuracy: ", np.mean([training_acc[0][-1],training_acc[1][-1], training_acc[2][-1], training_acc[3][-1], training_acc[4][-1] ]))
    print("Average Validation accuracy: ", np.mean([validation_acc[0][-1],validation_acc[1][-1], validation_acc[2][-1], validation_acc[3][-1], validation_acc[4][-1] ]))
    print("Average Training loss: ", np.mean([training_loss[0][-1],training_loss[1][-1], training_loss[2][-1], training_loss[3][-1], training_loss[4][-1] ]))
    print("Average Validation loss: ", np.mean([validation_loss[0][-1],validation_loss[1][-1], validation_loss[2][-1], validation_loss[3][-1], validation_loss[4][-1] ]))

main()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
import cv2
from glob import glob
import os
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.metrics import classification_report
from sklearn import metrics
import pandas as pd

seed = 182510
np.random.seed(seed)
tf.random.set_seed(seed)

def procImages(dataDir, labelDictionary, width, height):
    
    x = [] # will store images as arrays
    y = [] # store labels
    # list folders in directory
    directory = os.listdir(dataDir)
     
    # for each folder (train and validation) 
    for label in directory:
        
        # add class label to label dictionary
        if label not in labelDictionary:
            labelDictionary[label] = len(labelDictionary)
        
        # create full path for image directory (append absolute path and 
        # image directory path)
        sourceImages = os.path.join(dataDir, label)
        images = os.listdir(sourceImages)
        # for each image in directory, 
        for image in images:
            
            # read the image from file, resize and add to a list
            full_size_image = cv2.imread(os.path.join(sourceImages, image))
            x.append(cv2.resize(full_size_image, (width,height), 
                                                         interpolation=cv2.INTER_CUBIC))
            # add the class label to y
            y.append(label)

    return np.array(x), np.array(y)


def plotAccLoss(H, NUM_EPOCHS):

    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.show()


def main():
    # dimensions of our images.
    width, height = 64, 64
    K_fold = 5
    training_acc  = []
    validation_acc = []
    K_precision = []
    K_recall = []
    K_fscore = []
    K_validation_acc = []

    for i in range(K_fold):
        print('\nExecuting Fold: '+ str(i+1))
        print('Current Seed: ', np.random.get_state()[1][0])
        picFolderName = "Fold"+ str(i+1)
        train_data_dir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_training_set',picFolderName)
        validation_data_dir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_test_set',picFolderName)
        numTrainingSamples = 572
        numValidationSamples = 140
        NUM_EPOCHS = 50
        batchSize = 35

        classifier = Sequential()

        classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Conv2D(64, (3, 3), activation = 'relu'))
        classifier.add(MaxPooling2D(pool_size = (2, 2)))
        classifier.add(Flatten())
        classifier.add(Dense(units = 256, activation = 'relu', kernel_regularizer=regularizers.l2(l=0.02)))
        # classifier.add(Dropout(0.2))
        classifier.add(Dense(units = 128, activation = 'relu', kernel_regularizer=regularizers.l2(l=0.02)))
        # classifier.add(Dropout(0.2))
        classifier.add(Dense(units = 10, activation = 'softmax')) # number of classes

        classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

        # This dictionary will contain all labels with an associated int value
        labelDictionary ={}
        # The procImages function will read all image data from a folder
        # convert it to a Numpy array and add to a list
        # It will also add the label for the image, the folder name
        trainImages, trainLabels = procImages(train_data_dir, labelDictionary, width, height)
        valImages, valLabels = procImages(validation_data_dir, labelDictionary, width, height)

        # Map string label values to integer values. 
        trainLabels = (pd.Series(trainLabels).map(labelDictionary)).values
        valLabels = (pd.Series(valLabels).map(labelDictionary)).values

        trainLabels_values = np.max(trainLabels) + 1 
        trainLabels_onehot = np.eye(trainLabels_values)[trainLabels]
        valLabels_values = np.max(valLabels) + 1
        valLabels_onehot = np.eye(valLabels_values)[valLabels]

        train_datagen = ImageDataGenerator(
              rescale=1. / 255,
              shear_range=0.2,
              zoom_range=0.2,
              rotation_range = 20,
              horizontal_flip=True)

        train_generator = train_datagen.flow(
              trainImages, trainLabels_onehot, batch_size=batchSize, seed = seed)

        test_datagen = ImageDataGenerator(rescale=1. / 255)


        validation_generator = test_datagen.flow(valImages, valLabels_onehot, batch_size=batchSize, shuffle = False, seed = seed)

        history = classifier.fit(
            train_generator,
            steps_per_epoch=numTrainingSamples // batchSize,
            epochs=NUM_EPOCHS,
            validation_data=validation_generator,
            validation_steps=numValidationSamples // batchSize)
        
        training_acc.append(history.history["accuracy"])
        validation_acc.append(history.history["val_accuracy"])

        plotAccLoss(history, NUM_EPOCHS)

        classifier_json= classifier.to_json()
        with open("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp_extra.Fold_"+str(i+1)+"json", "w") as json_file:
              json_file.write(classifier_json)
          
        # serialize the model weights to a HDF5 file
        classifier.save_weights("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp_extra.Fold_"+str(i+1)+".h5")

        val_trues = valLabels
        validation_generator.reset()
        y_pred = classifier.predict(validation_generator, steps = numValidationSamples // batchSize)
        val_preds = np.argmax(y_pred, axis=1)

        labels = labelDictionary.keys()
        report = classification_report(val_trues, val_preds, output_dict=True)
        print(classification_report(val_trues, val_preds, target_names = labels))
        K_precision.append(report['weighted avg']['precision'])
        K_recall.append(report['weighted avg']['recall'])
        K_fscore.append(report['weighted avg']['f1-score'])
        K_validation_acc.append(report['accuracy'])
        print(metrics.confusion_matrix(val_trues, val_preds))

    print('\n')
    print('Average precision: ', np.mean(K_precision))
    print('Average recall: ', np.mean(K_recall))
    print('Average fscore: ', np.mean(K_fscore))
    print('Report accuracy: ', np.mean(K_validation_acc))
    print("Average Training accuracy: ", np.mean([training_acc[0][-1],training_acc[1][-1], training_acc[2][-1], training_acc[3][-1], training_acc[4][-1] ]))
    print("Average Validation accuracy: ", np.mean([validation_acc[0][-1],validation_acc[1][-1], validation_acc[2][-1], validation_acc[3][-1], validation_acc[4][-1] ]))


main()
```
