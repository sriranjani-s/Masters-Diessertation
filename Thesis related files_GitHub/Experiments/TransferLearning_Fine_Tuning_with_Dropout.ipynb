## MobileNet - Fine Tuning



```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import regularizers
import numpy as np
import cv2
import os
from glob import glob
import pandas as pd
from sklearn.metrics import classification_report
from sklearn import metrics
import random

seed = 182510
random.seed(seed)
tf.random.set_seed(seed)

def procImages(dataDir, labelDictionary, width, height):
    
    x = [] # will store images as arrays
    y = [] # store labels
    # list folders in directory
    directory = os.listdir(dataDir)
     
    # for each folder (train and validation) 
    for label in directory:
        
        # add class label to label dictionary
        if label not in labelDictionary:
            labelDictionary[label] = len(labelDictionary)
        
        # create full path for image directory (append absolute path and 
        # image directory path)
        sourceImages = os.path.join(dataDir, label)
        images = os.listdir(sourceImages)
        # for each image in directory, 
        for image in images:
            
            # read the image from file, resize and add to a list
            full_size_image = cv2.imread(os.path.join(sourceImages, image))
            x.append(cv2.resize(full_size_image, (width,height), 
                                                         interpolation=cv2.INTER_CUBIC))
            # add the class label to y
            y.append(label)

    return np.array(x), np.array(y)


def plotAccLoss(H, NUM_EPOCHS):

    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.show()



def main():
  # resize all images to thie width and height
    width, height= 64, 64
    training_acc = []
    validation_acc = []
    training_loss = []
    validation_loss = []
    K_precision = []
    K_recall = []
    K_fscore = []
    K_validation_acc = []

    for i in range(5):
        print('\nExecuting Fold'+str(i+1))  
        print('Current Seed: ', np.random.get_state()[1][0])
        picFolderName = 'Fold'+str(i+1)
        trainDataDir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_training_set',picFolderName)
        validationDataDir= os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_test_set',picFolderName)
        numTrainingSamples = 572
        numValidationSamples = 140
        NUM_EPOCHS = 90
        batchSize = 35

      # This dictionary will contain all labels with an associated int value
        labelDictionary ={}
      # The procImages function will read all image data from a folder
      # convert it to a Numpy array and add to a list
      # It will also add the label for the image, the folder name
        trainImages, trainLabels = procImages(trainDataDir, labelDictionary, width, height)
        valImages, valLabels = procImages(validationDataDir, labelDictionary, width, height)

        # Map string label values to integer values. 
        trainLabels = (pd.Series(trainLabels).map(labelDictionary)).values
        valLabels = (pd.Series(valLabels).map(labelDictionary)).values

        trainLabels_values = np.max(trainLabels) + 1
        trainLabels_onehot = np.eye(trainLabels_values)[trainLabels]
        valLabels_values = np.max(valLabels) + 1
        valLabels_onehot = np.eye(valLabels_values)[valLabels]
        # print(trainLabels_onehot, valLabels_onehot)

        train_datagen = ImageDataGenerator(
            rescale=1. / 255,
            shear_range=0.2,
            zoom_range=0.2,
            rotation_range = 20,
            horizontal_flip=True)

        train_generator = train_datagen.flow(
            trainImages, trainLabels_onehot, batch_size=batchSize, seed=seed, shuffle = False)

        test_datagen = ImageDataGenerator(rescale=1. / 255)


        validation_generator = test_datagen.flow(
            valImages, valLabels_onehot, batch_size=batchSize, shuffle = False, seed= seed)

        MobileNetModel = tf.keras.applications.MobileNet(weights='imagenet', include_top=False, input_shape=(64, 64, 3))

        MobileNetModel.trainable = False
        model = tf.keras.models.Sequential()
        model.add(MobileNetModel)
        model.add(tf.keras.layers.Flatten())
        model.add(tf.keras.layers.Dense(256, activation='relu'))  
        model.add(tf.keras.layers.Dropout(0.2))
        # model.add(tf.keras.layers.Dense(128, activation='relu'))  
        # model.add(tf.keras.layers.Dropout(0.2))
        model.add(tf.keras.layers.Dense(10, activation='softmax'))
        print (model.summary())

        print ("\n Phase A - Training Fully Connected Layers\n")
        model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.RMSprop(lr=0.001),
                  metrics=['accuracy'])
        H =model.fit(train_generator, 
                  steps_per_epoch=numTrainingSamples// batchSize,  
                  epochs=NUM_EPOCHS, batch_size=batchSize,       
                  validation_data=validation_generator,
                  validation_steps=numValidationSamples// batchSize,
                  shuffle = False)

        plotAccLoss(H, NUM_EPOCHS)

        print ("\n Phase B  - Fine Tune Fully Connected Layer and Selected Convolutional Layers \n")
        MobileNetModel.trainable = True
        trainableFlag = False
        for layer in MobileNetModel.layers:
          # print(layer.name)
            if layer.name == 'conv_dw_11':
                trainableFlag = True
            layer.trainable = trainableFlag
        model.compile(loss='categorical_crossentropy',
                    optimizer=tf.keras.optimizers.RMSprop(lr=1e-5),
                    metrics=['accuracy'])
        H =model.fit(train_generator, 
                  steps_per_epoch=numTrainingSamples// batchSize,  
                  epochs=NUM_EPOCHS, batch_size=batchSize,       
                  validation_data=validation_generator,
                  validation_steps=numValidationSamples// batchSize)

        plotAccLoss(H, NUM_EPOCHS)

        training_acc.append(H.history["accuracy"])
        validation_acc.append(H.history["val_accuracy"])
        training_loss.append(H.history["loss"])
        validation_loss.append(H.history["val_loss"])

        classifier_json= model.to_json()
        with open("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp24.Fold_"+str(i+1)+"json", "w") as json_file:
            json_file.write(classifier_json)

        # serialize the model weights to a HDF5 file
        model.save_weights("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp24.Fold_"+str(i+1)+".h5")

        val_trues = valLabels
        validation_generator.reset()
        y_pred = model.predict(validation_generator, steps = numValidationSamples // batchSize)
        val_preds = np.argmax(y_pred, axis=1)

        # labels = validation_generator.class_indices.keys()
        # print(labels)
        report = classification_report(val_trues, val_preds, output_dict=True, zero_division= 0)
        print(classification_report(val_trues, val_preds, zero_division = 0))
        K_precision.append(report['weighted avg']['precision'])
        K_recall.append(report['weighted avg']['recall'])
        K_fscore.append(report['weighted avg']['f1-score'])
        K_validation_acc.append(report['accuracy'])
        print(metrics.confusion_matrix(val_trues, val_preds))

    print('\n')
    print('Average precision: ', np.mean(K_precision))
    print('Average recall: ', np.mean(K_recall))
    print('Average fscore: ', np.mean(K_fscore))
    print('Report accuracy: ', np.mean(K_validation_acc))
    print("Average Training accuracy: ", np.mean([training_acc[0][-1],training_acc[1][-1], training_acc[2][-1], training_acc[3][-1], training_acc[4][-1] ]))
    print("Average Validation accuracy: ", np.mean([validation_acc[0][-1],validation_acc[1][-1], validation_acc[2][-1], validation_acc[3][-1], validation_acc[4][-1] ]))
    print("Average Training loss: ", np.mean([training_loss[0][-1],training_loss[1][-1], training_loss[2][-1], training_loss[3][-1], training_loss[4][-1] ]))
    print("Average Validation loss: ", np.mean([validation_loss[0][-1],validation_loss[1][-1], validation_loss[2][-1], validation_loss[3][-1], validation_loss[4][-1] ]))


main()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import regularizers
import numpy as np
import cv2
import os
from glob import glob
import pandas as pd
from sklearn.metrics import classification_report
from sklearn import metrics
import random

seed = 182510
random.seed(seed)
tf.random.set_seed(seed)

def procImages(dataDir, labelDictionary, width, height):
    
    x = [] # will store images as arrays
    y = [] # store labels
    # list folders in directory
    directory = os.listdir(dataDir)
     
    # for each folder (train and validation) 
    for label in directory:
        
        # add class label to label dictionary
        if label not in labelDictionary:
            labelDictionary[label] = len(labelDictionary)
        
        # create full path for image directory (append absolute path and 
        # image directory path)
        sourceImages = os.path.join(dataDir, label)
        images = os.listdir(sourceImages)
        # for each image in directory, 
        for image in images:
            
            # read the image from file, resize and add to a list
            full_size_image = cv2.imread(os.path.join(sourceImages, image))
            x.append(cv2.resize(full_size_image, (75,75), 
                                                         interpolation=cv2.INTER_CUBIC))
            # add the class label to y
            y.append(label)

    return np.array(x), np.array(y)


def plotAccLoss(H, NUM_EPOCHS):

    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.show()



def main():
  # resize all images to thie width and height
    width, height= 64, 64
    training_acc = []
    validation_acc = []
    training_loss = []
    validation_loss = []
    K_precision = []
    K_recall = []
    K_fscore = []
    K_validation_acc = []

    for i in range(5):
        print('\nExecuting Fold'+str(i+1))  
        print('Current Seed: ', np.random.get_state()[1][0])
        picFolderName = 'Fold'+str(i+1)
        trainDataDir = os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_training_set',picFolderName)
        validationDataDir= os.path.join('/content/gdrive/My Drive/Colab Notebooks/OpenPose_80_20_split_test_set',picFolderName)
        numTrainingSamples = 572
        numValidationSamples = 140
        NUM_EPOCHS = 90
        batchSize = 35

        # This dictionary will contain all labels with an associated int value
        labelDictionary ={}
        # The procImages function will read all image data from a folder
        # convert it to a Numpy array and add to a list
        # It will also add the label for the image, the folder name
        trainImages, trainLabels = procImages(trainDataDir, labelDictionary, width, height)
        valImages, valLabels = procImages(validationDataDir, labelDictionary, width, height)

        # Map string label values to integer values. 
        trainLabels = (pd.Series(trainLabels).map(labelDictionary)).values
        valLabels = (pd.Series(valLabels).map(labelDictionary)).values

        trainLabels_values = np.max(trainLabels) + 1
        trainLabels_onehot = np.eye(trainLabels_values)[trainLabels]
        valLabels_values = np.max(valLabels) + 1
        valLabels_onehot = np.eye(valLabels_values)[valLabels]
        # print(trainLabels_onehot, valLabels_onehot)

        train_datagen = ImageDataGenerator(
            rescale=1. / 255,
            shear_range=0.2,
            zoom_range=0.2,
            rotation_range = 20,
            horizontal_flip=True)

        train_generator = train_datagen.flow(
            trainImages, trainLabels_onehot, batch_size=batchSize, seed = seed)

        test_datagen = ImageDataGenerator(rescale=1. / 255)


        validation_generator = test_datagen.flow(
            valImages, valLabels_onehot, batch_size=batchSize, shuffle = False, seed = seed)

        InceptionModel = tf.keras.applications.InceptionV3(include_top=False, input_shape=(75,75,3))

        InceptionModel.trainable = False
        model = tf.keras.models.Sequential()
        model.add(InceptionModel)
        model.add(tf.keras.layers.Flatten())
        model.add(tf.keras.layers.Dense(256, activation='relu'))  
        model.add(tf.keras.layers.Dropout(0.2))
        # model.add(tf.keras.layers.Dense(128, activation='relu'))  
        # model.add(tf.keras.layers.Dropout(0.2))
        model.add(tf.keras.layers.Dense(10, activation='softmax'))
        print (model.summary())

        print ("\n Phase A - Training Fully Connected Layers\n")
        model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.RMSprop(lr=0.001),
                  metrics=['accuracy'])
        H =model.fit(train_generator, 
                  steps_per_epoch=numTrainingSamples// batchSize,  
                  epochs=NUM_EPOCHS, batch_size=batchSize,       
                  validation_data=validation_generator,
                  validation_steps=numValidationSamples// batchSize)

        plotAccLoss(H, NUM_EPOCHS)

        print ("\n Phase B  - Fine Tune Fully Connected Layer and Selected Convolutional Layers \n")
        InceptionModel.trainable = True
        trainableFlag = False
        for layer in InceptionModel.layers:
          # print(layer.name)
            if layer.name == 'conv2d_166':
                trainableFlag = True
            layer.trainable = trainableFlag
        model.compile(loss='categorical_crossentropy',
                    optimizer=tf.keras.optimizers.RMSprop(lr=1e-5),
                    metrics=['accuracy'])
        H =model.fit(train_generator, 
                  steps_per_epoch=numTrainingSamples// batchSize,  
                  epochs=NUM_EPOCHS, batch_size=batchSize,       
                  validation_data=validation_generator,
                  validation_steps=numValidationSamples// batchSize)

        plotAccLoss(H, NUM_EPOCHS)

        training_acc.append(H.history["accuracy"])
        validation_acc.append(H.history["val_accuracy"])
        training_loss.append(H.history["loss"])
        validation_loss.append(H.history["val_loss"])

        classifier_json= model.to_json()
        with open("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp28.Fold_"+str(i+1)+"json", "w") as json_file:
          json_file.write(classifier_json)

        # serialize the model weights to a HDF5 file
        model.save_weights("/content/gdrive/My Drive/Colab Notebooks/Project Experiments_V1/Saved_models_V2/classifier_exp28.Fold_"+str(i+1)+".h5")

        val_trues = valLabels
        validation_generator.reset()
        y_pred = model.predict(validation_generator, steps = numValidationSamples // batchSize)
        val_preds = np.argmax(y_pred, axis=1)

        # labels = validation_generator.class_indices.keys()
        # print(labels)
        report = classification_report(val_trues, val_preds, output_dict=True, zero_division= 0)
        print(classification_report(val_trues, val_preds, zero_division = 0))
        K_precision.append(report['weighted avg']['precision'])
        K_recall.append(report['weighted avg']['recall'])
        K_fscore.append(report['weighted avg']['f1-score'])
        K_validation_acc.append(report['accuracy'])
        print(metrics.confusion_matrix(val_trues, val_preds))

    print('\n')
    print('Average precision: ', np.mean(K_precision))
    print('Average recall: ', np.mean(K_recall))
    print('Average fscore: ', np.mean(K_fscore))
    print('Report accuracy: ', np.mean(K_validation_acc))
    print("Average Training accuracy: ", np.mean([training_acc[0][-1],training_acc[1][-1], training_acc[2][-1], training_acc[3][-1], training_acc[4][-1] ]))
    print("Average Validation accuracy: ", np.mean([validation_acc[0][-1],validation_acc[1][-1], validation_acc[2][-1], validation_acc[3][-1], validation_acc[4][-1] ]))
    print("Average Training loss: ", np.mean([training_loss[0][-1],training_loss[1][-1], training_loss[2][-1], training_loss[3][-1], training_loss[4][-1] ]))
    print("Average Validation loss: ", np.mean([validation_loss[0][-1],validation_loss[1][-1], validation_loss[2][-1], validation_loss[3][-1], validation_loss[4][-1] ]))


main()
```
